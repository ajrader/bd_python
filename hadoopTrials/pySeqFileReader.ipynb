<h3> Now push infile into a dictionary: keys are the fis and values are the counts</h3>

```python
reader = SequenceFile.Reader(infile)
key_class = reader.getKeyClass()
value_class = reader.getValueClass()
key = key_class()
value = value_class()
fisDict = {}
position = reader.getPosition()
while reader.next(key,value):
#    print '*' if reader.syncSeen() else ' ',
#    print '[%6s] %6s %6s ' % (position, key.toString(), value.toString())
    fiskey = frozenset(key.toString().split())
    fisDict[fiskey] = int(value.toString())
    position = reader.getPosition()
    
close(infile)
```

```python
import re
fptxtfile = 'myFP3.txt'
#KVpattern = re.compile('Key:\s(\d+):\sValue:\s([.*])')#[((?:\\d+\\s*)+)]\\s+=>\\s+{((?:\\d+\\s*)+)}\\s+([0-9.]+)%$')
fispatterns = {}
with open(fptxtfile,'r') as fo:
    for line in fo:
        if line.startswith('Key:'):
            #kv = line.split(':')
            #print len(kv), kv[0],kv[1],kv[2],kv[3]
            #keyid = kv[1]
            #values = kv[3].split('),')
            #for val in values:
            #    print val
            m = re.match(r'Key: (\d+): Value: ((.*))',line)
            #m = re.match( r'Key:\\s(\\d+):\\sValue:\\s([.*])',line,re.M|re.I)#KVpattern.match(line)
            #print m.group()
            if m is None: 
                print "Error with line:",line
                break
            else:
                keyid = m.group(1)
                values = m.group(2)
                fispatterns[keyid] = []
                for vv in values.split('),'):
                    vvv = vv.split('],')
                    occ = vvv[1]
                    try:
                        occ = int(occ)
                    except ValueError:
                        occ = int(occ[:-1])
                    
                    items = []
                    for a in vvv[0].split('[')[1].split(','):
                        items.append(int(a))
                        #print keyid,':',occ,'\t', int(a)
                    fispatterns[keyid].append((items,occ))
```
